More info about the dataset columns in kaggle.

*Note*: you can find the dataset in this repo: countries_of_the_world.csv

```python
# Open the dataset from the local filesystem
df_countries = pd.read_csv('../../data/countries_of_the_world.csv')
```

```python
# Select columns Country, Region and Population
df_countries[['Country', 'Region', 'Population']]
```

```python
df_countries.loc[:, ['Country', 'Region', 'Population']]
```

```python
# Select the 4 first rows of the dataframe, columns Country, Population and Area
df_countries.loc[0:3, ['Country', 'Population', 'Region']]
```

```python
df_countries.head(4)[['Country', 'Population', 'Region']]
```

```python
df_countries.iloc[0:4, [0, 1, 2]]
```

```python
# Show column names
df_countries.columns.tolist()
```

```python
# Select the last 6 rows of the dataset, columns: country, literacy and Phones
df_countries.loc[df_countries.index[-6:], ['Country', 'Literacy (%)', 'Phones (per 1000)']]
```

```python
df_countries.tail(6).loc[:, ['Country', 'Literacy (%)', 'Phones (per 1000)']]
```

```python
# Select the two firts rows using loc or iloc
df_countries.iloc[-6:, [0, df_countries.columns.get_loc('Literacy (%)'), 
                        df_countries.columns.get_loc('Phones (per 1000)')
                    ]
                ]
```

```python
# Select the fifth row
df_countries.iloc[[4]]
```

```python
# Select data from Algeria (using loc)
df_countries.loc[df_countries['Country'] == 'Algeria ']
```

```python
# Save data about Albania in a new dataframe variable
# Then show the new dataframe
df_albania = df_countries.query("Country == 'Algeria '")
df_albania
```

```python
# Create a new dataframe calle datos_10_paises 
# columns: Country, Population and Climate 
# rows: the ten first ones
# Then show the new dataframe
data = df_countries.loc[:9, ['Country', 'Population', 'Climate']]
data.to_csv('../../data/datos_10_paises.csv', index=False)
```

```python
# Load the dataframe again using country names as index
df_10_countries = pd.read_csv('../../data/datos_10_paises.csv')
df_10_countries
```

```python
# Show 5 random rows
df_10_countries.sample(5)
```

```python
#Show data about France (use indexes)
df_countries.set_index('Country', inplace = True)

df_countries.loc[['France ']]

# Note: If you have problems with indexes, take a look to the CSV file directly ;-)
```

```python
# Create a new dataframe with data about Spain
# Columns: Population, Area, Climate, Birthrate and Deathrate
df_spain = df_countries.loc[['Spain '], ['Population', 'Climate', 'Birthrate', 'Deathrate']]
df_spain
```

```python
# Select all data from these countries: France, Germany and Spain
df_countries.loc[['France ', 'Germany ', 'Spain ']]
```

```python
# Select a range of rows
# Select rows from 10 to 20 (not included)
df_countries.iloc[10:20]
```

```python
# Select rows from Spain to Uruguay
# Show only Population and Density columns
df_countries.loc['Spain ':'Uruguay ', ['Population', 'Pop. Density (per sq. mi.)']]
```
